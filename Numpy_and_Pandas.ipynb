```python
#  Question 1. importing  CSV file
import pandas as pd
df = pd.read_csv("/content/retail_dataset .csv")
df
df.head()
df.info()
```

```python
# Question 2. find missing value in netitre dataset.
total_missing = df.isnull().sum()

print("Total missing values in the dataset:", total_missing)
```

```python
# Queastion 3. fill missing value with mean
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].mean())
df["Price"] = df["Price"].fillna(df["Price"].mean())
df
```

```python
# Question 4 removing row where exist missing value. category and region. dropna ke sath subset use hota hai.
df = df.dropna(subset=["Product Category", "Region"])
df
```

```python
# Question 5. Create a new column called Revenue using NumPy
import numpy as np
df["Revenue"] = np.multiply(df["Quantity"], df["Price"])
df
```

```python
#Q6. Using NumPy, calculate the total revenue generated from all transactions in the dataset.
total_revenue = np.sum(df["Revenue"])
print(f"total_revenue generated from all transaction is {total_revenue}")
```

```python
# Q7. Group the dataset by Product Category and calculate the total revenue generated by each category.
revenue_by_category = df.groupby("Product Category")["Revenue"].sum().round(2)
print(revenue_by_category)
```

```python
# Question 8 Top 3 highest revenue and bottom 3 lowest revenue by category revenue
Category_revenue = df.groupby("Product Category")["Revenue"].sum().round(2)
Top_3 = Category_revenue.sort_values(ascending=False).head(3)
Bottom_3 = Category_revenue.sort_values().head(3)
print("Top 3 highest revenue by category:")
print(Top_3)
print("\nBottom 3 lowest revenue by category:")
print(Bottom_3)
```

```python
# Questin 9 region wise revenue and display min and maximum revenue by which region
Region_Revenue = df.groupby("Region")["Revenue"].sum()
print(Region_Revenue)
print("Region with the highest revenue:", Region_Revenue.idxmax(), "-" , Region_Revenue.max())
print("Region with the lowest revenue:", Region_Revenue.idxmin(), "- ", Region_Revenue.min())
```

```python
# Question
import numpy as np

# Convert Date column into datetime
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)

# Extract Month
df["Month"] = df["Date"].dt.month

# Monthly Revenue
monthly_revenue = df.groupby("Month")["Revenue"].sum()

print("Monthly Revenue:")
print(monthly_revenue)

# Mean Revenue
print("\nMean Revenue:", np.mean(df["Revenue"]))

# Median Revenue
print("Median Revenue:", np.median(df["Revenue"]))

# Standard Deviation
print("Standard Deviation:", np.std(df["Revenue"]))
```
